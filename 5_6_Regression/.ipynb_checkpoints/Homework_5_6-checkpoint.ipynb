### I.III Target analysis

```python
data_target = data[data[target]!=-200][target] # we analyse target variable without missings, as we will drop them later
sns.histplot(data_target, kde=True) 
```

We can observe positive skewness of the target's distribution. In the next steps we need to scale it because those rare, big values can have a bad impact on our model's performance. But firstly, let's try to find a transformation of this variable which will give us a distribution as close to normal as possible.

```python
sns.histplot(np.sqrt(data_target), kde=True).set_title('Sqrt and ^1/3 target distribution', size=13)# lambda = 1/2
sns.histplot(np.power(data_target,1/3), kde=True, color='green') # lambda = 1/3
```

The second option (raising variable to 1/3 power) seems to be a good one - we have a distribution close to normal. However, our data is positive so we can use ***Box-cox*** transformation to determine the best *lambda* to get normal distribution.

```python
original_data = data_target
fitted_data, fitted_lambda = boxcox(original_data) 
fig, ax = plt.subplots(1, 2, sharey = True, figsize=(10,5)) 
  
sns.histplot(original_data, kde = True, ax = ax[0]).set_title('Original data', size=14)
sns.histplot(fitted_data,  kde = True, color ="green", ax = ax[1]).set_title('Data after transformation', size=14) 
ax[1].set(xlabel = 'C6H6(GT)^0.27')
print(f"Optimal lambda value: {fitted_lambda}") 
```

As we can see after raising our target to 0.27 power we get rid of the skewness and get quite good approximation of normal distribution.

## II. Data preparation & feature engineering

### II.I Dropping missing values & outliers

From the EDA we know that *NMHC(GT)* is mostly missing (90% of it) so we will drop this variable, because gives us no information. We can also see that there are probably 366 records missing, with only *Date* and *Time* values provided. They also won't be informative so we will drop them. As for the ~14% missings left in *CO2*, *NO2*, *NOx* - we will impute them as we cannot afford dropping so many records.

```python
# checking whether the same records
data.loc[(data['PT08.S1(CO)'] == -200) & (data['PT08.S2(NMHC)'] == -200) & (data['PT08.S3(NOx)'] == -200) 
          & (data['PT08.S4(NO2)'] == -200) & (data['PT08.S5(O3)'] == -200) & (data['C6H6(GT)']==-200)].shape
```

```python
data = data.drop('NMHC(GT)', axis=1).loc[data['PT08.S1(CO)'] != -200].reset_index(drop=True) 
data.shape
```

```python
# outliers identified earlier
data = data.loc[(data[target] < 60) & (data['PT08.S3(NOx)'] < 2200 )].reset_index(drop=True) 
data.shape
```

### II.II Feature engineering

We still have two categorical variables to deal with - *Date* and *Time*. We can parse them into one, *datetime* format variable and treat it as an order for our data, not a feature. Our data seems to be sorted already, but we will perform this operation just to make sure and drop the *Date* variable from our data. We could leave it as an index itself, but later on the numerical indices will be more useful to us.

```python
for i in range(len(data)):
    row = data.loc[i,'Date']
    data.loc[i,'Date'] = row.replace('/','-')
```

```python
data['Date'] = pd.to_datetime(data.Date+' '+data.Time, format='%d-%m-%Y %H.%M.%S')
```

```python
data.Date
```

```python
data = data.sort_values('Date',axis=0).drop(['Date','Time'], axis=1)
```

```python
data.info()
```

### II.III Scaling and imputation

We will just define the strategy here, it will be implemented later after we split our data. For those three variables with missings we will perform median imputation, combined with scaling of our data. We choose median over mean because it has better statistical properties. All of our data is numerical, so we just need to exclude target variable from our transformations.

```python
num_cols = data.select_dtypes('number').columns.to_list()
num_cols.remove(target)
```

```python
numerical_vars = gen_features(columns=[[col] for col in num_cols],
                              classes=[{'class': SimpleImputer, 'strategy': 'median', 'missing_values': -200},
                                  {'class': StandardScaler}])

preprocessor = DataFrameMapper(numerical_vars) 
```

## IV. Baseline model

### IV.I Train-test split

For the first fitting of our model we will split our data into two datasets with ratio 80/20. We do not shuffle the data, as we want to preserve our sorting by date and time.

```python
seed=123
```

```python
data_target = data[target]
data_train, data_val, target_train, target_val, = train_test_split(data.drop(target, axis=1), data_target, 
                                                                   random_state=seed, shuffle=False, test_size = 0.2)
```

```python
#print(data_train.shape, data_val.shape, target_train.shape, target_val.shape)
```

```python
train = pd.DataFrame(preprocessor.fit_transform(data_train))
val = pd.DataFrame(preprocessor.transform(data_val))

train_y = pd.Series(target_train)
val_y = pd.Series(target_val)
```

### IV.II Model evaluation

```python
linear_reg = LinearRegression()
linear_reg.fit(train, train_y)
print(f'\nR2 Score: {linear_reg.score(val, val_y)}')
```

R2 score looks good, but as we know, it can be misleading. We don't know if the score is so high due to predictive power of our model, or is it just because of the number of variables we have in our dataset. Let's define some more metrics to verify that result.

### IV.III Metrics

* MAE - Mean Absolute Error - minimized by the median, by which we impute the data
* MSE - Mean Squared Error - minimized by mean, penalizing large errors
* RMSE - Root Mean Squared Error - square root of MSE
* R2 - R-squared - the proportion of the variance in the dependent variable that is predictable from the independent variables
* Adj_R2 - Adjusted R-squared; modified version of R-squared that adjusts for the number of predictors in a regression model

```python
model = Pipeline([
    ('preprocessor', preprocessor),
    ('estimator', LinearRegression())
])
```

```python
def evaluate(X_train, X_val, Y_train, Y_val, pipeline):
    pipeline.fit(X_train, Y_train)
    pred_val = pipeline.predict(X_val)
    pred_train = pipeline.predict(X_train)
    # preprocessor = pipeline.named_steps['preprocessor']
    
    return pd.DataFrame({
        'MAE': [mean_absolute_error(Y_train, pred_train), mean_absolute_error(Y_val, pred_val)], 
        'MSE': [mean_squared_error(Y_train, pred_train), mean_squared_error(Y_val, pred_val)],
        'RMSE': [np.sqrt(mean_squared_error(Y_train, pred_train)), np.sqrt(mean_squared_error(Y_val, pred_val))],
        'R2': [r2_score(Y_train, pred_train), r2_score(Y_val, pred_val)],
        'Adj R2': [1 - (1-r2_score(Y_train, pred_train))*(len(Y_train)-1)/(len(Y_train)-X_train.shape[1]-1),
                   1 - (1-r2_score(Y_val, pred_val))*(len(Y_val)-1)/(len(Y_val)-X_val.shape[1]-1)]
    }, index=['Train','Val']) , preprocessor.transformed_names_, pred_val
```

```python
metrics, colnames, pred = evaluate(data_train, data_val, target_train, target_val, model)
metrics
```

We can see that both R2 and Adj R2 are really high, but the errors doesn't indicate a good performance of our model. Maybe there are some outliers present, which cause such results. We can also see much higher errors for our validation set and lower R-squared - means we're dealing with overfitting.

### IV.IV Residual analysis

```python
def plot_res(pred_val, Y_val):
    
    errors = Y_val - pred_val
    fig, axes = plt.subplots(1,2,figsize=(12,6))
    sns.scatterplot(list(range(len(errors))), errors, ax=axes[0]).set_title("Distibution of errors vs index")
    axes[0].set(ylabel='Error', xlabel='Index')

    sns.scatterplot(pred, errors, ax=axes[1]).set_title("Distibution of errors vs fitted value")
    axes[1].set(ylabel='Error', xlabel='Fitted value')
```

```python
plot_res(pred, target_val)
```

We can see that our errors aren't uniformly distributed, therefore we have a heteroscedasticity problem - variance of the errors differ depending on the value of our prediction. A solution to that is e.g. Box-Cox transformation, which we evaluated on the beginning - it helps us identify the true relationship between our data and target (raising to power 0.27 turned out to be the best modification). We will analyse the impact of this change in a moment, but firstly let's drop the outliers visible on the plot.

```python
pred = model.predict(data_val)
errors = target_val - pred
out = errors[errors>5].index #to_list()
```

```python
normal_index = [index for index in data_val.index if index not in out]
data_val = data_val.loc[normal_index]
target_val = target_val.loc[normal_index]
```

We can see that those three features are very highly correlated, but before dropping them we should check how important they are in our model.

### IV.VI Feature importance

```python
def plot_importance(model, colnames):
    importance = pd.DataFrame({'Weight': np.abs(model.coef_),'Variable': colnames})
    imp = importance.sort_values(by='Weight', ascending=False)
    sns.barplot( x='Weight',y='Variable', data=imp).set_title('Feature importance', size=14)
```

```python
plot_importance(model.named_steps['estimator'], colnames)
```

Both features correlated with the one on the top have small contribution to our model, so we can drop them, alongside with the RH variable, which we earlier identified as highly correlated with AH, T variables.

```python
top_cor.append('RH')
high_cor = [f for f in numerical_vars if f[0][0] in top_cor]
pipeline = Pipeline([
    ('mapper', DataFrameMapper(high_cor)),
    ('estimator', LinearRegression())
])
```

```python
metrics, colnames, pred = evaluate(data_train, data_val, target_train, target_val, pipeline)
metrics
```

As we can see dropping those features didn't help. Let's now try a different approach. 

## V. Model with Box-Cox transformation

As we already noticed, we need to deal with heteroscedasticity problem, and one way to do so is to transform our target variable. During EDA phase we identified the optimal lambda value which results in the best approximation of a normal distribution curve.

```python
fig, ax = plt.subplots(1, 2, sharey = True, figsize=(10,5)) 
sns.histplot(data_target, kde = True, ax = ax[0]).set_title('Original data', size=14)
sns.histplot(data_target**0.27,  kde = True, color ="green", ax = ax[1]).set_title('Data after transformation', size=14) 
ax[1].set(xlabel = 'C6H6(GT)^0.27')
```

Let's know evaluate how the performance of our model improves after transforming the target variable.

```python
model = Pipeline([
    ('preprocessor', preprocessor),
    ('estimator', LinearRegression())
])
```

```python
metrics, _, pred = evaluate(data_train, data_val, np.power(target_train,0.27), np.power(target_val,0.27), model)
metrics
```

As we can see above, errors are significantly lower than before. There's still a difference in the results between train and validation set, so probably we could still improve the model. However, results above are satisfactory.

## VI. Tuning hyperparameters (cross-validation)

Now we will try different approach - replacing basic linear regression with a regularized one (Ridge/Lasso/ElasticNet). We will penalize for too big number of features in our model.

### VI.I TimeSeriesSplit

To perform crossvalidation to find best parameters for our model we need to adequately split our data. As we deal with time series data, we cannot shuffle it because we will loose important information.

```python
Y = data_target
X = data.drop(target,axis=1)
tscv = TimeSeriesSplit(gap=0, n_splits=5)

for train_index, test_index in tscv.split(data):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = Y.loc[train_index], Y.loc[test_index]
```

### VI.II Lasso regression

```python
lasso = Lasso()
model = Pipeline([('preprocessor', preprocessor), 
                     ('estimator', lasso)])
grid = {
    'estimator__alpha': np.linspace(0, 300, 5)
}
gs = GridSearchCV(model, grid, 
                  n_jobs=-1,
                  scoring=['neg_mean_squared_error', 'neg_mean_absolute_error'], 
                  refit='neg_mean_squared_error', 
                  cv=tscv)
```

```python
gs.fit(data, data_target)
gs.best_params_
```

```python
pipeline = clone(model)
pipeline.set_params(**gs.best_params_);

scores, colnames, pred = evaluate(data_train, data_val, target_train, target_val, pipeline)
scores
```

There's no regularization (alpha=0) so we get same results as for the normal linear regression. It simply means it is not optimal for us to drop any variables form our data (as lasso estimator works like a feature selection - zeros the coefficients).
